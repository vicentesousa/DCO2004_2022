```matlab
fig = figure;

Ns = 10^3;                                             % Número de símbolos simulados
snrs = [0 5 20];
markers = ['ro'; 'bs'; 'gp'; 'c*'; 'm.'; 'yx'; 'kd'; 'c^'];
constelations = [2 8];

for j = 1:length(constelations)
    M = constelations(j);
    for i = 1:length(snrs)
        EbN0_dB = snrs(i);
        % Gráficos
        subplot(length(snrs), length(constelations), j+(i-1)*length(constelations));
        bits = randi([0 M-1], Ns, 1);
        txSig = pskmod(bits, M);
        rxSig = awgn(txSig, EbN0_dB);
        legends = [];
        % Gráficos
        for m = 1:M
            plot(rxSig(find(bits == m-1)), markers(m, :));
            legends = [legends; ['Símbolo ' num2str(m-1) ' transmitido']];
            hold on;
        end

        title(['Diagrama de ' num2str(M) ' constelações com EbN0 = ' num2str(EbN0_dB) 'dB']);
        legend(legends, 'Location', 'BestOutside');
        minAx = min([real(rxSig)' imag(rxSig)']);
        maxAx = max([real(rxSig)' imag(rxSig)']);
        axis([minAx maxAx minAx maxAx]);
    end
end
fig.PaperUnits = 'inches';
fig.PaperPosition = [0 0 16 4*length(snrs)];
```

Para ambos casos, temos que um aumento na SNR distancia as constelações e, portanto, torna mais fácil distingui-las.

Não obstante, um aumento de constelações  tem um efeito inverso e reduz a distância entre constelações vizinhas. Contudo, o aumento do número de constelações permite um número maior de bits transmitidos.

Sendo assim, podemos facilmente observar a relação de compromisso entre o número de constelações, a probabilidade de erro e a taxa de transmissão do sistema.
