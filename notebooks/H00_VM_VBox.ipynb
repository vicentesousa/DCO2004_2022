## Criação de máquina virtual Ubuntu Virtual box (UBUNTU 20.04)
## Objetivos
As metas desse tutorial são ajudar o usuário a:
- Instalar e configurar uma maquina virtual Linux Ubuntu 20.04 LTS (Focal Fossa).


**Passo 1**: Baixe e instale o VirtualBox em seu computador (para acessar a página de Download [acesse aqui](https://www.virtualbox.org/).

![Virtual1](../FIGS/H00_virtualbox/virtual_box.png)

**Passo 2**: Baixe a iso do Ubuntu 20.04-3 LTS (Focal Fossa).

- Versão 64 Bits: [https://ubuntu.com/download/desktop](https://ubuntu.com/download/desktop)

**Passo 3**: Após instalar o Virtualbox e baixar a iso, abra o VirtualBox e clique em **Novo**, e depois na opção **Criar** de acordo com a imagem:

![fig_Virtual2](../FIGS/H00_virtualbox/virtual_box1.png)

**Passo 4**: Escolha a memória que deseja alocar, o ideal é que seja a metade da memória de sua máquina:

![fig_Virtual2](../FIGS/H00_virtualbox/virtual-box-memoria.png)

**Passo 5**: Nessa tela escolha a opção **VDI (VirtualBox Disk Image)** e prossiga a instalção:

![fig_VM2](../FIGS/H00_virtualbox/virtual_box2.png)

**Passo 6**: Escolha a opção **Dinamicamente Alocado**:

![fig_VM3](../FIGS/H00_virtualbox/virtual_box3.png)

**Passo 7**: Informe o nome do disco virtual e especifique **60 GB** de espaço em HD (se tiver pouco espaço em HD, pode criar uma máquina de **40 GB**):

![fig_VM3](../FIGS/H00_virtualbox/virtual_box4.png)

**Passo 8**: Inicialize a máquina criada com dois cliques sobre ela:

![fig_VM3](../FIGS/H00_virtualbox/virtual_box5.png)

**Passo 9**: Localize a iso do ubuntu baixada no ícone da pasta:

![fig_VM3](../FIGS/H00_virtualbox/virtual_box6.png)

**Passo 10**: Selecione a iso, para começar o processo de instalação:

![fig_VM3](../FIGS/H00_virtualbox/virtual_box7.png)

**Passo 11**: Espere carregar a máquina e continuar o processo de instalação:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu1.jpg)

**Passo 12**: Aguarde enquanto carrega:

![fig_VM3](../FIGS/H00_virtualbox/inicio_ubuntu.jpg)

**Passo 13**: Escolha o idoma para prosseguir selecione a opção Install para prosseguir a instalação:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu2.jpg)

**Passo 14**: Escolha o idioma do teclado:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu4.jpg)

**Passo 15**: Nas opções abaixo configure de acordo com a imagem:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu6.jpg)

**Passo 16**: escolha a primeira opção e continue a instalação:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu7.jpg)

**Passo 17**: Nessa tela click na opção **continue**:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu8.jpg)

**Passo 18**: Informe sua localização e click em **continue** novamente:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu9.jpg)


**Passo 19**: Preencha as informações,referentes a máquina como desejar:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu10.jpg)

**Passo 20**: Aguarde enquanto é feita instalção:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu11.jpg)

**Passo 21**: Espere carregar a máquina e continuar o processo de instalação: 

![fig_VM3](../FIGS/H00_virtualbox/ubuntu12.jpg)

**Passo 22**: Assim que concluída escolha a opção **reset**:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu14.jpg)

**Passo 23**: Sua máquina virtual está pronta para uso:

![fig_VM3](../FIGS/H00_virtualbox/ubuntu15.jpg)

**Passo 24 (muito importante)**: Instale o VirtualBox Guest Additions. Em um terminal, digite:

sudo apt install build-essential dkms linux-headers-$(uname -r)

Depois sigas as instruções de [instalação](https://www.tecmint.com/install-virtualbox-guest-additions-in-ubuntu/#:~:text=VirtualBox%20Guest%20Additions%20are%20a,and%20usability%20of%20guest%20systems.&text=Easy%20mouse%20pointer%20integration.).

Depois de reiniciar sua máquina virtual, você verá uma menu com várias opções, entre elas a de maximizar o VirtualBox para ocupar toda a tela. Ao maximizar, o mesmo menu aparecerá no canto inferior.

![fig_VM3](../FIGS/H00_virtualbox/menu_vbox.png)
